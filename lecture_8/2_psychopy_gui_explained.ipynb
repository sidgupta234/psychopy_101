# PsychoPy `gui.Dlg` Complete Explanation Notebook
This notebook explains all features of `psychopy.gui.Dlg` with examples.
Includes checkboxes, dropdowns, text fields, required fields, enabling/disabling fields, tooltips, and capturing output.

```python
from psychopy import gui, core
import pandas as pd
import numpy as np
```

## Basic Dialog

```python
textBox = gui.Dlg(title='Basic Example')
textBox.addField('First Name')
textBox.addField('Last Name')
textBox.addField('Age')
show_dlg = textBox.show()
show_dlg
```

## Checkbox Example

```python
dlg = gui.Dlg(title='Checkbox Example')
dlg.addField('Subscribe to newsletter?', initial=False)
dlg.show()
```

## Dropdown Example

```python
dlg = gui.Dlg(title='Dropdown Example')
dlg.addField('Choose difficulty', choices=['Easy','Medium','Hard'])
dlg.show()
```

## Required Fields

```python
dlg = gui.Dlg(title='Required Example')
dlg.addField('Participant ID', required=True)
dlg.addField('Session Number')
dlg.show()
```

## Tooltips & Colors

```python
dlg = gui.Dlg(title='Style Example')
dlg.addField('Name', tip='Enter your full name')
dlg.addField('Secret Code', color='red')
dlg.show()
```

## Disabled Fields

```python
dlg = gui.Dlg(title='Disabled Example')
dlg.addField('Auto-filled value', initial='1234', enabled=False)
dlg.show()
```

## Capturing Dialog Output

```python
dlg = gui.Dlg(title='Output Capture Example')
dlg.addField('Name')
dlg.addField('Age')
results = dlg.show()
print('Results:', results)
```

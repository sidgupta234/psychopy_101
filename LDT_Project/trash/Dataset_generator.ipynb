```python
import nltk
import random
nltk.download('wordnet')
```

```python
def shuffle(word):
    letters_list = list(word)
    random.shuffle(letters_list)
    return ''.join(letters_list)

def shuffle_except_first_last(word):
    letters_list = list(word)[1:-1]
    random.shuffle(letters_list)
    return word[0] + ''.join(letters_list) + word[-1]
```

```python
from nltk.corpus import wordnet
```

```python
word1 = ["tea", "bread", "doctor", "truck", "lava" , "cloth", "fruit", "engine", "office", "guitar", "insect", "blood", "photo", "shirt", "tongue", "chocolate", "hotel", "virus", "phone", "tennis"]
word2 = ["coffee", "butter", "nurse", "vehicle", "magma", "apparel", "orange", "train", "chair", "violin", "termite", "platelet", "camera", "trouser", "muscle", "biscuit", "lobby", "bacteria", "radio", "soccer"]
word3 = ["bicycle", "laptop", "shampoo", "paper", "biscuit", "charger", "airplane", "skeleton", "jelly", "lemon", "lunch", "skirt", "burger", "flower", "pencil", "diary", "onion", "bottle", "light", "mouse"]
wup_1_2 = []
wup_1_3 = []
word2_shuffle_list = []
word2_shuffle_except_first_last_list = []

for i in (word1, word2, word3):
     print(len(i))
```

```python
for i in word3:
    print(i)
```

```python

for i in range(0, len(word1)):
    #print(word2[i])
    syn1 = wordnet.synsets(word1[i])[0]
    syn2 = wordnet.synsets(word2[i])[0]
    syn3 = wordnet.synsets(word3[i])[0]
    wup_1_2_value = syn1.wup_similarity(syn2)
    wup_1_3_value = syn1.wup_similarity(syn3)
    wup_1_2.append(round(wup_1_2_value, 2))
    wup_1_3.append(round(wup_1_3_value, 2))

for i in range(len(wup_1_2)):
    print(wup_1_2[i], word1[i])

print("")

for i in range(len(wup_1_3)):
    print(wup_1_3[i], word3[i])
```

```python
for i in range(0, len(word2)):
    word2_shuffle = shuffle(word2[i])
    word2_shuffle_except_first_last = shuffle_except_first_last(word2[i])
    word2_shuffle_list.append(word2_shuffle)
    word2_shuffle_except_first_last_list.append(word2_shuffle_except_first_last)
```

```python
print(word2_shuffle_list)
print(word2_shuffle_except_first_last_list)
```

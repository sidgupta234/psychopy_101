```python
list_of_colors = ['yellow' for i in range(8)] 
dataset.boxplot(column = ['reaction_time'], by=['condition', 'response_status'], showmeans=True,
               patch_artist = True)#color=list_of_colors)#['reaction_time']
#axes.findobj(axes.patches).set_facecolor("red")
```

```python
list_of_colors = ['yellow' for i in range(8)] 
bp_dict = dataset.boxplot(column = ['reaction_time'], by=['condition', 'response_status'], showmeans=True,
               patch_artist = True, color='black', return_type='both')#color=list_of_colors)#['reaction_time']
#axes.findobj(axes.patches).set_facecolor("red")
#colors = ['blue', 'green', 'purple', 'tan', 'pink', 'red', 'grey', 'yellow']

colors = ['#999999', 'yellow', '#999999', 'yellow', '#999999', 'yellow', '#999999', 'yellow']
#print(len(bp_dict.iteritems()))
for row_key, (ax,row) in bp_dict.iteritems():
    #ax.set_xlabel('')
    for i,box in enumerate(row['boxes']):
        box.set_facecolor(colors[i])

plt.suptitle('')
plt.xlabel("Conditions and Responses")
plt.ylabel("Reaction Time")
#plt.xticks(['a', 'b', 'c', 'd'])
x_labels =['(1) Correct', '(1) Incorrect', 
           '(2) Correct', '(2) Incorrect',
          '(3) Correct', '(3) Incorrect',
          '(4) Correct', '(4) Incorrect']
plt.xticks(range(1,9), x_labels, rotation=30)
#plt.yticks(range(0,110, 10))
# plt.bar(age_distribution.index, age_distribution)
plt.legend(["RT distribution"])
plt.title("Reaction time across various conditions and responses")
plt.show()

plt.show()
#dataset.plot(kind='box',color='blue')
```
